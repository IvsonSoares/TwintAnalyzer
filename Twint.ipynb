```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import punkt
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
import spacy
```

```python
def LimpaLinha(string):
    
    #removing mentions and hashtags
    string = ' '.join(re.sub(r'(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\s+)','' , string).split())
    string = ' '.join(re.sub(r'(#[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\s+)','' , string).split())
    
    #removing links
    
    string = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', ' ', string, flags=re.MULTILINE)
    
    #removing words and caracters
    
    stop_words = set(stopwords.words('portuguese'))
    word_tokens = word_tokenize(string)
    filtered_sentence = [w for w in word_tokens if not w in stop_words]
    filtered_sentence = []
    
    for w in word_tokens:
        if w not in stop_words and len(w) > 3:
            filtered_sentence.append(" "+w)
            
            
    string = ''.join(filtered_sentence)
    
    return string.lower()
```

```python

df_raw['clean_text'] = df_raw.apply(lambda row: LimpaLinha(row.tweet), axis = 1)
```
